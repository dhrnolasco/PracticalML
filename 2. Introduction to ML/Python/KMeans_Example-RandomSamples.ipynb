## Unsupervised Learning Example: KMeans

This example uses KMeans Algorithm in sklearn library to cluster a set of samples

```python
#load required functions and libraries
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline


#prepare the data
data=pd.read_csv("xclara.csv")#load csv data file
print(data.head())
plt.scatter(data[["V1"]],data[["V2"]],marker='.')
```

```python

#convert data to samples
X=np.array(data.values)
#apply KMeans to K clusters
ClustersCount=3
kmeans = KMeans(n_clusters=ClustersCount)
kmeans.fit(X)#fit the samples, and find the clusters

print("Clusters Centroids")
for a,b in enumerate(kmeans.cluster_centers_):
    print("{0} --> {1}".format(a,b))
```
