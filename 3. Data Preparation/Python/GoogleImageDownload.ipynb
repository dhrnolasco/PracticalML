## Getting Data from Google Images

This notebook shows how to use existing search services, such as google images, to gather a dataset of images which you can use in your application.

For this to work, we will use "google_images_download" library to run it in Python.

You will need to download google chromedriver from here (select the one matches your operating system):

```python
from google_images_download import google_images_download as gd

#make sure to modify this one to point to the extract file 
chromedriver_location="/Users/yamen/Downloads/chromedriver"

def download_images(images,count,output_dir="download"):
    response = gd.googleimagesdownload()
    arguments = {"keywords":",".join(images),"limit":count,"print_urls":True,"output_directory":output_dir,
                 "chromedriver":chromedriver_location}
    paths=response.download(arguments)
    return paths
```

```python
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

#write down the categories you want to search for
keywords=["rabbit","cat","dog","mouse","tiger","duck"]
count_per_keyword=200
output_dir="animals"

#this function will start downloading the images to the output_dir folder
paths=download_images(keywords,count_per_keyword,output_dir=output_dir)
```

```python
import os 
from PIL import Image

def create_folder(path):
    if(os.path.exists(path)):
        return path
    os.mkdir(path)
    return path


def process_image(img):
    width=64
    height=64
    img2=img.resize((width,height))
    return img2

labels=os.listdir(output_dir)
X_train=[]
Y_train=[]


out_dir=create_folder("output")

for p in labels:
    if not os.path.isdir(os.path.join(output_dir,p)):
        continue
    files=os.listdir(os.path.join(output_dir,p))
    label_folder=create_folder(os.path.join(out_dir,p))
    for i,f in enumerate(files):
        try:
            img=Image.open(os.path.join(output_dir,p, f))
            processed_img=process_image(img)
            fname=os.path.join(label_folder,"img_{0}.jpg".format(i))
            processed_img.save(fname)
            processed_img.close()
            img.close()
        except:
            continue
        
    #get some statisitcs about the processed images
    count=len(os.listdir(os.path.join(output_dir,p)))
    print("Total processed images for [{0}] is {1}",p,count)

```
